###<font color='#ff471a'>What is Numpy?</font>
Numpy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays. The ancestor of NumPy, Numeric, was originally created by Jim Hugunin with contributions from several other developers. In 2005, Travis Oliphant created NumPy by incorporating features of the competing Numarray into Numeric, with extensive modifications. NumPy is open-source software and has many contributors.

###<font color='#0000e6'>By using the - operator:</font>


```python
# For example, a[0:3] means the first 2 rows (rows 0 through 1 (include)) of a.
a[0:2]
```

```python
#For example, a[0:3] means the first 3 rows (rows 0 through 2) of a.
a[0:3]
```

###<font color='#0000e6'>By using the negative operator</font>


```python
#For example, a[-1] means the last row of x.
print(a[-1])
```

###<font color='#0000e6'>By using the : operator</font>


```python
#For example, a[2:] means the rest of the rows+(include row 2) after row 2.
a[2:]
```

```python
#Similarly, a[:2] means all the rows up to row 2 (not include row 2).
a[:2]
```

```python
#a[:] means all rows of a.
a[:]
```

###<font color='#0000e6'>The sort Function:</font>

###<font color='red'>numpy.sort(a, axis=- 1, kind=None, order=None)</font>

  Input:
*  A: The array to Sort
*  Axis:Axis along which to sort. If None, the array is flattened before sorting. The default is -1, which sorts along the last axis. 
*   Kind: quicksort/ heapsort/ mergesort/ timesort
*   order: When a is an array with fields defined, this argument specifies which fields to compare first, second, etc. A single field can be specified as a string

Output:


*   Return a sorted copy of an array.



```python
 a = np.array([[4, 3, 5], [1, 2, 1]])	
a
```

rt1.JPG

```python
  #axis=-1: returns an array,flat the array to one big array and sorted it!
 print(np.sort(a, axis=-1) )
```

```python
 # axis=1:returns an array,sorted by rows!
 print(np.sort(a, axis=1) )
```

```python
 # axis=0:returns an array,sorted by columns!
 print(np.sort(a, axis=0))
```
